# 4. Write a function with name Goal_diff_count which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

# Goal_diff_count = GoalsFor - GoalsAgainst

```python
df["GoalsFor"] = df["GoalsFor"].astype(int)
df["GoalsAgainst"] = df["GoalsAgainst"].astype(int)
```

```python
# Goal_diff_count(): 
# This function will calculate goal difference of each team. 
# It will also print team which in having maximum and minimun goal difference
# Return: This function will return dataframe which has team name and goal difference. Apart from these it will also return 
# Max and mini goal difference team

def Goal_diff_count(dataFrame):
    df_goal_diff = dataFrame.filter(["Team"], axis=1)
    df_goal_diff["Goal Difference"] = dataFrame["GoalsFor"] - dataFrame["GoalsAgainst"]
    
    #team which has maximum goal difference
    max_goal_diff = df_goal_diff["Goal Difference"].max()
    Maxteam = df_goal_diff[df_goal_diff["Goal Difference"] == max_goal_diff]
    print("Team with Maximun Goal Difference:")
    print("Team: {}, Goal Difference: {}".format(Maxteam.iloc[0]["Team"], Maxteam.iloc[0]["Goal Difference"]))
    print("\n")
    
    #team which has minimun goal difference
    min_goal_diff = df_goal_diff["Goal Difference"].min()    
    Minteam = df_goal_diff[df_goal_diff["Goal Difference"] == min_goal_diff]
    print("Team with Minimun Goal Difference:")
    print("Team: {}, Goal Difference: {}".format(Minteam.iloc[0]["Team"], Minteam.iloc[0]["Goal Difference"]))
    
    return df_goal_diff, Maxteam.iloc[0]["Team"], Maxteam.iloc[0]["Goal Difference"],Minteam.iloc[0]["Team"], Minteam.iloc[0]["Goal Difference"]
```

```python
GoalDiff_df, TeamA, Diff_A, TeamB, Diff_B = Goal_diff_count(df)
```

### Team with maximum and minimun goal difference can also be printed from "GoalDiff_df" dataframe, as shown below.

```python
#team which has maximum goal difference
print("Team with Maximun Goal Difference:")
max_goal_diff = GoalDiff_df["Goal Difference"].max()
GoalDiff_df[GoalDiff_df["Goal Difference"] == max_goal_diff]
```
